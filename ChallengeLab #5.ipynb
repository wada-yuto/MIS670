# Example 1: Simple Word Cloud

```python
from os import path
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt
%matplotlib inline

text = "one two three four five six seven eight nine ten eleven"

text2 = "one two three four five six seven eight nine ten eleven eleven eleven eleven eleven"

# Generate a word cloud image
wordcloud = WordCloud().generate(text2)

# Display the generated image:
# the matplotlib way:
plt.imshow(wordcloud)
plt.axis("off")

# take relative word frequencies into account, lower max_font_size
wordcloud = WordCloud(max_font_size=40, relative_scaling=.5).generate(text2)
plt.figure()
plt.imshow(wordcloud)
plt.axis("off")
```

# Example 4:  Using a mask you can generate wordclouds in arbitrary shapes

The example below shows how to create a word cloud on a mask image. For the mast image, you need to use **png** (not jpg) files. 
Some exemplar png files are in the examples folder inside the data folder. You can download additional image files (stencils) from such websites as http://www.freestencilgallery.com/category/animals/ and https://www.template.net/design-templates/print/printable-stencil-template/. It appears that the images on these websites are in jpg format. You need to download and save them as **png** file (e.g., **filename.png**). You can also download image files from Google Image.

Nice tutorial on using mask for word clouds
- http://minimaxir.com/2016/05/wordclouds/
- http://fontawesome.io/
- https://github.com/Pythonity/icon-font-to-png

```python
# Example 4a: Masked Word Cloud
# ================
# Using a mask you can generate wordclouds in arbitrary shapes. Bigfoot Text and Data
data_folder = Path("data") # Notice PC Users that slashes are backwards!
file_to_read = data_folder / "bigfoot2.txt"
mask_to_read = data_folder / "bigfoot2.png"
file_to_write1 = data_folder / "bigfoot_msk.png"
file_to_write2 = data_folder / "bigfoot_msk.pdf"


from os import path
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
from wordcloud import WordCloud, STOPWORDS

# Read the whole text.
text = open(file_to_read, encoding="utf8").read()

stopwords = set(STOPWORDS)
stopwords.add("im")

# read the mask image
bigfoot_mask = np.array(Image.open(mask_to_read))

wc = WordCloud(background_color="white", max_words=2000, mask=bigfoot_mask,
               stopwords=stopwords)
# generate word cloud
wc.generate(text)

# show
plt.figure(figsize=(30,30))
plt.imshow(wc)
plt.axis("off")
plt.savefig(file_to_write1)
plt.savefig(file_to_write2)

plt.figure()
plt.imshow(bigfoot_mask, cmap=plt.cm.gray)
plt.axis("off")
```

```python
# Example 4b: Image-colored word cloud: Bigfoot text and colored Squatch Mask 
# ========================
# You can color a word-cloud by using an image-based coloring strategy implemented in
# ImageColorGenerator. It uses the average color of the region occupied by the word
# in a source image. You can combine this with masking - pure-white will be interpreted
# as 'don't occupy' by the WordCloud object when passed as mask.
# If you want white as a legal color, you can just pass a different image to "mask". 
# We will pass black.


from os import path
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator

file_to_read = data_folder / "bigfoot2.txt"
mask_to_read = data_folder / "colorsquatch.jpg"
file_to_write1 = data_folder / "bigfoot_mskc.png"
file_to_write2 = data_folder / "bigfoot_mskc.pdf"

# Read the whole text.
text = open(file_to_read, encoding="utf8").read()

# read the mask / color image
# https://www.teeshirtpalace.com/products/trippy-sasquatch-tie-dye-colorful-bigfoot-poster
bf_coloring = np.array(Image.open(mask_to_read))
stopwords = set(STOPWORDS)
stopwords.add("said")

wc = WordCloud(background_color="black", max_words=3000, mask=bf_coloring,
               stopwords=stopwords, max_font_size=40, random_state=42)
# generate word cloud
wc.generate(text)

# create coloring from image
image_colors = ImageColorGenerator(bf_coloring)

# show
plt.imshow(wc)
plt.axis("off")


# recolor wordcloud and show
# we could also give color_func=image_colors directly in the constructor
plt.figure(figsize=(16,16))
plt.imshow(wc.recolor(color_func=image_colors))
plt.axis("off")
plt.savefig(file_to_write1)
plt.savefig(file_to_write2)

plt.figure()
plt.imshow(bf_coloring, cmap=plt.cm.gray)
plt.axis("off")
```

```python
# Example 4c: Using bigrams and from_frequencies from Researchers at Stanford (Bigfoot with Batman Shape)
# ==================================
# We are using a custom tokenizer (here implemented from scratch, it's recommended to
# use nltk, spacy or scikit-learn instead), to allow the inclusion of word-pairs
# (bigrams, 2-grams) into the word cloud.
# The ``from_frequencies`` method allows generating wordclouds from a list or
# array of ``(word, frequency)`` tuples, where ``word`` can be any string, and
# ``frequency`` can be any int or float.
# We are using the likelihood ratio score developed by Dunning to find "collocations",
# which are phrases made up of two or more words (we only consider two here).
# If the chance that a bigram is a collocation is high, we discount the appearances
# of the single words -- otherwise they would always be at least as big as the bigram.

import numpy as np
from PIL import Image
from os import path
import matplotlib.pyplot as plt
import random
from itertools import tee
from collections import defaultdict
import re
from math import log
from wordcloud import WordCloud, STOPWORDS

from pathlib import Path
data_folder = Path("data")
file_to_read = data_folder / "bigfoot2.txt"
mask_to_read = data_folder / "batman.png"
file_to_write1 = data_folder / "bigfoot_mskbm.png"
file_to_write2 = data_folder / "bigfoot_mskbm.pdf"


# dunning's likelihood ratio with notation from
# http://nlp.stanford.edu/fsnlp/promo/colloc.pdf


def l(k, n, x):
    return log(max(x, 1e-10)) * k + log(max(1 - x, 1e-10)) * (n - k)


def score(bigram, counts, n_words):
    N = n_words
    c12 = counts[bigram]
    c1 = counts[bigram[0]]
    c2 = counts[bigram[1]]
    p = c2 / N
    p1 = c12 / c1
    p2 = (c2 - c12) / (N - c1)
    score = l(c12, c1, p) + l(c2 - c12, N - c1, p) - l(c12, c1, p1) - l(c2 - c12, N - c1, p2)
    return -2 * score


def grey_color_func(word, font_size, position, orientation, random_state=None, **kwargs):
    return "hsl(0, 0%%, %d%%)" % random.randint(60, 100)


def pairwise(iterable):
    # from itertool recipies
    # is -> (s0,s1), (s1,s2), (s2, s3), ...
    a, b = tee(iterable)
    next(b, None)
    return list(zip(a, b))


def unigrams_and_bigrams(text, stopwords=None):
    stopwords = [s.lower() for s in stopwords] if stopwords is not None else []
    words = re.findall(r"\w[\w']+", text)
    # remove stopwords
    words = [word for word in words if word.lower() not in stopwords]
    # remove 's
    words = [word[:-2] if word.lower().endswith("'s") else word for word in words]
    # fix for movie-script upper case names
    words = [word if not word.isupper() else word.title() for word in words]
    n_words = len(words)
    # make tuples of two words following each other
    bigrams = list(pairwise(words))
    counts_unigrams = defaultdict(int)
    counts_bigrams = defaultdict(int)
    for word in words:
        counts_unigrams[word] += 1
    for bigram in bigrams:
        # join tuples by a space
        counts_bigrams[bigram] += 1

    counts_all = {}
    counts_all.update(counts_unigrams)
    counts_all.update(counts_bigrams)

    # decount words inside bigrams
    for bigram in list(counts_bigrams.keys()):
        # collocation detection (30 is arbitrary):
        if score(bigram, counts_all, n_words) > 30:
            counts_unigrams[bigram[0]] -= counts_bigrams[bigram]
            counts_unigrams[bigram[1]] -= counts_bigrams[bigram]
        # add joined bigram into unigrams
        counts_unigrams[' '.join(bigram)] = counts_bigrams[bigram]
    return counts_unigrams


# Read the whole text.
text = open(file_to_read, encoding="utf8").read()


# read the mask image
mask = np.array(Image.open(mask_to_read))


from IPython.display import Image, display
display(Image(filename="data/batman.png"))


wc = WordCloud(max_words=1000, mask=mask, margin=10,
               color_func=grey_color_func, random_state=3)
# from_freqencies ignores "stopwords" so we have to do it ourselves
wc.generate_from_frequencies(unigrams_and_bigrams(text, STOPWORDS))
plt.figure(figsize=(30,30))
plt.imshow(wc)
wc.to_file(file_to_write1)
plt.axis("off")

```
